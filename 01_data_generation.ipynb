```python
import math
import pandas as pd
import numpy as np
from scipy.stats import norm, bernoulli, uniform, expon
from scenarios import *
from utils import *
```

```python
seed = 0xC0FFEE
test_seed = 0xDECAF
n = 10_000
```

```python
rng = np.random.default_rng(seed)


X1 = norm.rvs(size=n, loc=0, scale=1, random_state=rng)
X2 = bernoulli.rvs(size=n, p=0.5, random_state=rng)
X3 = norm.rvs(size=n, loc=0, scale=1, random_state=rng)
X4 = uniform.rvs(size=n, loc=-2, scale=4, random_state=rng)
X5 = expon.rvs(size=n, scale=1, random_state=rng)
X6 = rng.choice(['A','B','C'], size=n, p=[1/3]*3)
X7 = norm.rvs(size=n, loc=0, scale=1, random_state=rng)
X8 = rng.choice(['0','1'], size=n, p=[1/2]*2)
X9 = norm.rvs(size=n, loc=0, scale=1, random_state=rng)

eps_1 = norm.rvs(size=n, loc=0, scale=1, random_state=rng)
eps_4 = norm.rvs(size=n, loc=0, scale=2, random_state=rng)
eps_9 = norm.rvs(size=n, loc=0, scale=3, random_state=rng)

base_frame = pd.DataFrame(dict(
    X1=X1,
    X2=X2,
    X3=X3,
    X4=X4,
    X5=X5,
    X6=X6,
    X7=X7,
    X8=X8,
    X9=X9,
))

###
header("X1, X3, X7, X9 ~ N(0, 1)")
print(f"mean(X1) = {float(np.mean(X1)):.4f}; var(X1) = {float(np.var(X1)):.4f}")
print(f"mean(X3) = {float(np.mean(X3)):.4f}; var(X3) = {float(np.var(X3)):.4f}")
print(f"mean(X7) = {float(np.mean(X7)):.4f}; var(X7) = {float(np.var(X7)):.4f}")
print(f"mean(X9) = {float(np.mean(X9)):.4f}; var(X9) = {float(np.var(X9)):.4f}")
header("eps_1, eps_4, eps_9 ~ N(0, {1,4,9})")
print(f"mean(eps_1) = {float(np.mean(eps_1)):.4f}; var(eps_1) = {float(np.var(eps_1)):.4f}")
print(f"mean(eps_4) = {float(np.mean(eps_4)):.4f}; var(eps_4) = {float(np.var(eps_4)):.4f}")
print(f"mean(eps_9) = {float(np.mean(eps_9)):.4f}; var(eps_9) = {float(np.var(eps_9)):.4f}")
header("X2 ~ Bernoulli(0.5)")
print(f"mean(X2) = {float(np.mean(X2)):.4f}")
header("X4 ~ U(-2,2)")
print(f"mean(X4) = {float(np.mean(X4)):.4f}, min(X4) = {float(np.min(X4)):.4f}, max(X4) = {float(np.max(X4)):.4f}")
header("X5 ~ Exp(1)")
print(f"mean(X5) = {float(np.mean(X5)):.4f}")
header("X6 in {A, B, C}")
print(f"X6 == 'A': {np.mean(X6=='A')}; X6 == 'B': {np.mean(X6=='B')}; X6 == 'C': {np.mean(X6=='C')}")
header("X8 in {0, 1}")
print(f"X8=='0': {np.mean(X8 == '0')}; X8=='1': {np.mean(X8 == '1')}")

base_frame
```

```python
Y_1_base = gt_scenario_1(X1,X2,X3,X4,X5,X6,X7,X8,X9)

Y_1_1 = Y_1_base + eps_1
Y_1_4 = Y_1_base + eps_4
Y_1_9 = Y_1_base + eps_9

base_frame.assign(Y=Y_1_1).to_csv("data/synth_1_1.csv", index=False)
base_frame.assign(Y=Y_1_4).to_csv("data/synth_1_4.csv", index=False)
base_frame.assign(Y=Y_1_9).to_csv("data/synth_1_9.csv", index=False)
```
